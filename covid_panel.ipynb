```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
N = 300
covid_data = pd.read_csv("./data/MICRODADOS.csv", encoding="latin-1", sep = ';', parse_dates=['DataDiagnostico', 'DataObito'])
covid_data
```

```python
confirmed = covid_data[covid_data["Classificacao"] == "Confirmados"]
city_cases = confirmed.groupby(by="Municipio").size().to_frame(name="Casos Confirmados")
confirmed_count = city_cases.sort_values(by="Casos Confirmados", ascending=False)

confirmed
```

```python
# print(confirmed.columns)
city = "Vitoria"
confirmed_in_city = confirmed[confirmed["Municipio"] == city.upper()]
confirmed_by_date = confirmed_in_city.groupby(by="DataDiagnostico").size().to_frame("Qnt de casos").cumsum()
dead_by_date = confirmed_in_city.groupby(by="DataObito").size().to_frame("Qnt de mortes").cumsum()
# print(confirmed_by_date)
# print(dead_by_date)

x = confirmed_by_date

fig, my_axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 5))
fig.tight_layout()

my_ax_0 = sns.lineplot(data=confirmed_by_date, color="r", ax=my_axes[0])
my_ax_1 = sns.lineplot(data=dead_by_date, color="b", ax=my_axes[1])
```

```python
cities = ['Vitoria', 'Serra', "Vila Velha"]

for i in range(len(cities)):
    cities[i] = cities[i].upper()

mask_cities = confirmed["Municipio"].isin(cities)
confirmed_by_cities = confirmed[mask_cities]
confirmed_by_cities = confirmed_by_cities[["Municipio", "IdadeNaDataNotificacao"]]
confirmed_by_cities["IdadeNaDataNotificacao"] = confirmed_by_cities["IdadeNaDataNotificacao"].str.split(" anos").str[0].astype("int32")

confirmed_by_age = confirmed_by_cities.groupby(by=["Municipio", "IdadeNaDataNotificacao"]).size().to_frame("Qnt de casos")
print(confirmed_by_age)

fig = plt.subplots(figsize=(15, 5))

sns.scatterplot(data=confirmed_by_age, 
            x="IdadeNaDataNotificacao", 
            y="Qnt de casos",
            hue="Municipio",
            palette="Blues",
           )
```
